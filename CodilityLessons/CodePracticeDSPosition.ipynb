```python
import pandas as pd
import numpy as np

timestamp = ['2020-02-16 16:38:56', '2020-11-21 11:59:50',
'2020-08-13 12:54:24','2020-10-12 06:14:03',
'2020-06-13 02:54:24', '2020-08-13 12:54:24', '2020-11-21 11:59:50', '2020-08-13 12:54:24', '2020-12-19 00:44:47', '2019-01-01 04:30:07']

data = {'timestamp': timestamp,
'transient_1': [ 11, 62, 3, 62, 11, 3, 62, 62, 62, 11], 'transient_2' : [166,166, 0,225,225, 0,225, 0,225,300], 'transient_3' : [867, 83,903, 8, 29,105,424,455,410,437], 'transient_4': [ 94,756,927,191,611,117, 45,268,180,273], 'transient_5': [ 95,272,754, 5,477,749, 1,390,943, 4], 'transient_6' : [105,424,455,410,437, 45,268,180,550,273], 'transient_7': [112,201,355,342,495,135,689, 2,958,661], 'transient_8' : [163,735,848, 67,855,198,302,112,416,217], 'transient_9': [ 1,390,495,943, 4,408,163,661,484,943],}

df = pd.DataFrame(data=data)
```

```python
# Add to 'df' a column called 'average' which contains the average of the transients

df['average'] = df.loc[:, df.columns != 'timestamp'].mean(axis=1)
df
```

```python
# Add to 'df' a column called 'average_7_max' which contains the average of the largest 7 transients

df['average_7_max'] = df.loc[:, list(set(df.columns) - set(('timestamp' ,'average', 'max')))].apply(lambda x: np.mean(sorted(x.values.tolist())[-7:]), axis=1)
df
```

```python
# Add to 'df' a column called 'month' containing the month number (jan-> 1, feb ->2, ...) of the timestamp

df['month'] = df.timestamp.apply(lambda x: int(x.split('-')[1]))
df
```

```python
# In 'df', divide by 10 all the transient values between 200 and 800

for c in df.columns:
  if 'transient' in c:
    df[c] = np.where((df[c] >= 200) & (df[c] <= 800), df[c]/10, df[c])

df
```

```python
# In 'df' replace the transients  of row 7 by [1,2,3,4,5,6,7,8,9]

df.iloc[7, 1:10] = [1,2,3,4,5,6,7,8,9]
df
```

```python
# Sort 'df' by the value contained in timestamp column, in case of equal timestamp, the column transient 1 should be used to sort the rows, in case of identical timestamp and transient_1, the value transient_2 should be used to sort the rows... until transient 10

df = df.sort_values(by=list(df.columns[:10]), ignore_index=True)
```

```python
# Add to df the columns transient_diff_i = transient_i - transient_{i-1} for i = 2, ..., 9

for i in range(2,10):
  df[f'transient_diff_{i}'] = df[f'transient_{i}']-df[f'transient_{i-1}']
df
```

```python
# Multiply the transient values of row 1  by 1, of row 2 by 10, of row 3 by 100 etc.

for i, v in df.iterrows():
  df.iloc[i, 1:10] = df.iloc[i, 1:10]*(10**i)
  print(10**i)

df
```
